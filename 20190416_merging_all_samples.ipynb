```python
import pandas as pd
import os
```

```python
basedir = '/mnt/c/MinION16S_analysis/output'
```

```python
#This is getting the full path of all taxID.csv files
taxID_csv = [os.path.join(basedir, x) for x in os.listdir(basedir) if x.endswith('taxID.csv')]
```

```python
#I checked if I got the right amount of csv input files
len(taxID_csv)
```

```python
taxID_csv
```

```python
# [1:] means from second file to last
taxID_csv[1:]
```

```python
#assigns first file to the dataframe to initialise
merged_df = pd.read_csv(taxID_csv[0], sep=',')
```

```python
merged_df.head()
```

```python
#loop; csv_fn - all tax id files from 2 to 21; tmp_df reads all the files; merged_df merges all the files.
for csv_fn in taxID_csv[1:]:
    tmp_df = pd.read_csv(csv_fn, sep=',')
    merged_df = pd.merge(merged_df,tmp_df, how = 'outer', left_on = 'taxid', right_on = 'taxid' )
merged_df.fillna(value=0, inplace=True)
```

```python
#change index coloumn and drop taxid.
merged_df.index = merged_df.taxid
merged_df.drop('taxid', axis=1, inplace=True)
```

```python
merged_df.head()
```

```python
merged_df.columns
```

```python
#rearranging the coloumns
new_columns = ['Hare_1', 'Hare_2', 'Hare_3', 'Hare_4', 'Hare_5', 'Hare_6',
       'Hare_7', 'Hare_8', 'Hare_9', 
       'Rabbit_1', 'Rabbit_2', 'Rabbit_3', 'Rabbit_4', 'Rabbit_5', 'Rabbit_6',
       'Rabbit_7', 'Read_8', 'Rabbit_9', 'Rabbit_10', 'Rabbit_11', 'Rabbit_12']
```

```python
#ordering the table; starts with rows and then coloumns
merged_df = merged_df.loc[:, new_columns]
```
